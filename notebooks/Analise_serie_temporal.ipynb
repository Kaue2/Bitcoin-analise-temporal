# Análises

Nesse primeiro momento vamos fazer analises sobre toda a série temporal e também sobre pequenos recortes de interesse para o estudo, sendo esses recortes nos anos de: 2014, 2017, 2018, 2020, 2024 e 2025. 

As análises tem como objetivo explicar a sazonalidade, ruídos e tendências observadas a partir de médias móveis e dos dados originais, todas as médias foram cálculadas considerando janelas de 30 dias para suavizar os ruídos e ter maior clareza dos dados.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

btc = yf.download("BTC-USD", period="max")
#print(btc.head)
plt.plot(btc['Close'], label='Preço de Fechamento')
plt.title("Preço diário do Bitcoin (BTC-USD)")
plt.xlabel("Data")
plt.ylabel("Preço em USD")
plt.legend()
plt.show()
```

```python
btc['SMA_30'] = btc['Close'].rolling(window=30).mean()

plt.figure(figsize=(12, 6))
plt.plot(btc['Close'], label='Preço de fechamento', alpha=0.7)
plt.plot(btc['SMA_30'], label='Média móvel simples (30 dias)', linestyle='solid')
plt.legend()
plt.title('BTC-USD com médias móveis de 30 dias')
plt.show()
```

Podemos observar que a média móvel suaviza os ruídos dos dados, vamos observar como essa média móvel precede momentos de alta e de baixa no mercado

```python
momento_alta = btc[(btc.index >= "2017-07-01")
    & (btc.index <= "2018-06-30")].copy()
#print(momento_alta.head());
momento_alta['MediaAlta'] = momento_alta['Close'].rolling(window=30).mean()

plt.figure(figsize=(12, 6))
plt.plot(momento_alta['Close'], label='Preço de fechamento', alpha=0.7)
plt.plot(momento_alta['MediaAlta'],
         label='Média móvel simples (30 dias)',
         linestyle='solid')
plt.legend()
plt.title('BTC-USD com médias móveis de 30 dias')
plt.show()
```

Podemos observer atráves do gráfico que em momentos onde o valor de fechamento do Bitcoin é superior a média logo após temos uma tendência de que a média suba, o contrário também é valido uma vez que os valores a baixo da média afetam a média negativamente. Podemos observar que esse padrão do bitcoin se repte em outros momentos mais a frente, porém a tendência a longo prazo é de que o bitcoin suba.

```python
btc = yf.download("BTC-USD", start="2018-07-1", end="2020-06-1")
btc['media'] = btc['Close'].rolling(window=30).mean()

plt.figure(figsize=(12, 6))
plt.plot(btc['Close'], label='Preço de fechamento', alpha=0.7)
plt.plot(btc['media'],
         label='Média móvel simples (30 dias)',
         linestyle='solid')
plt.legend()
plt.title('BTC-USD com médias móveis de 30 dias')
plt.show()
```

Podemos notar novamente o mesmo padrão, um momento de alta seguido por um declive e depois a establiziação do mercado, isso se segue ao longo de toda linha temporal com exporadicos "boons" que são um aumento considerável de seu valor em média, como nos anos de 2018, final de 2021, metade de 2024 e 2025.

```python
btc = yf.download("BTC-USD", start="2014-01-1", end="2017-12-31")
btc['media'] = btc['Close'].rolling(window=30).mean()

plt.figure(figsize=(12, 6))
plt.plot(btc['Close'], label='Preço de fechamento', alpha=0.7)
plt.plot(btc['media'],
         label='Média móvel simples (30 dias)',
         linestyle='solid')
plt.legend()
plt.title('BTC-USD com médias móveis de 30 dias')
plt.show()
```

Nesse período podemos observar outro fenômeno que é a estabilidade do preço do bitcoin, algo que é bem incomum em períodos posteriores. A média móvel nos ajuda a observer esse fenômeno uma vez que os dados não estão muito a cima ou muito abaixo da linha da média móvel, isso significa que as tendências do mercado nesse ponto são incertas, logo a frente por volta de janeiro de 2017, podemos ver que os preços de fechamento do bitcoin passam consideravelmente da linha e isso já indica uma tendência de crescimento. Podemos observer que mais uma vez a sazonalidade se repete com um momento de declive logo após o aumento do bitcoin.
