```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

btc = yf.download("BTC-USD", period="max")
#print(btc.head)
plt.plot(btc['Close'], label='Preço de Fechamento')
plt.title("Preço diário do Bitcoin (BTC-USD)")
plt.xlabel("Data")
plt.ylabel("Preço em USD")
plt.legend()
plt.show()
```

```python
btc['SMA_30'] = btc['Close'].rolling(window=30).mean()

plt.figure(figsize=(12, 6))
plt.plot(btc['Close'], label='Preço de fechamento', alpha=0.7)
plt.plot(btc['SMA_30'], label='Média móvel simples (30 dias)', linestyle='solid')
plt.legend()
plt.title('BTC-USD com médias móveis de 30 dias')
plt.show()
```
